```python
#Encoding
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
df.drop(columns = ['order_month'],inplace=True)
```

```python
df.columns
```

```python
nominal_cols = ['type_of_vehicle','festival','city_type','order_day','is_weekend']
ordinal_cols = ['weather','traffic']
numeric_cols = ['ratings','vehicle_condition','multiple_deliveries','pickup_time_minutes','distance']
```

```python
df['traffic'].unique()
```

```python
seq_order = [['sunny','cloudy','windy', 'fog','sandstorms','stormy'],['low','medium','high','jam']]
```

```python
df
```

```python
Transformer = ColumnTransformer([('Nominal Encodding',OneHotEncoder(),nominal_cols),
                ("Ordinal Encoding",OrdinalEncoder(categories=seq_order),ordinal_cols),
                ("Feature Scaling",StandardScaler(),numeric_cols)])
```

```python
Transformer
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = Pipeline([('Data Preprocessing',Transformer),
          ('ML Algorithm',LinearRegression())])
```

```python
X = df.drop('time_taken',axis = 1)
y = df['time_taken']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Linear Regression
model_lr = Pipeline([('Data Preprocessing',Transformer),
          ('ML Algorithm',LinearRegression())])
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Linear Regression")
print("R2 Score",r2_score(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#Decision Tree
model = Pipeline([('Data Preprocessing',Transformer),
          ('ML Algorithm',DecisionTreeRegressor())])
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Decision Regression")
print("R2 Score",r2_score(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestRegressor

model = Pipeline([('Data Preprocessing',Transformer),
          ('ML Algorithm',RandomForestRegressor())])
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("RF Regression")
print("R2 Score",r2_score(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
```
